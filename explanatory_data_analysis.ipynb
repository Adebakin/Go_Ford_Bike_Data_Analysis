## Dataset Overview

The dataset consist of one hundred and eighty three thousnad four hundred and twelce(183412) rows(samples) measured over sixteen(16) features/variables/ columns. The Go Ford bike is a bike hiring services that matches a member with an electronic assisted bicycle to navigate the busy, in most cases heavy traffic roads so that computing time can be reduced and efficiency can be increased as the case may be. The bike has been noticed to climb hilly roads that could have been stressful on regular byclces.
* duration_sec : the diference between the time a person hires bike and returns it at the dock which is the end time.
* start_time : the start time the bicycle is hired
* end_time : the end time the byclec is returned to the dock
* start_station_id : the station id the bike is hired from
* start_station_name : the station name the bike to hired from
* start_station_latitude : the lattitude the bike is hired from
* start_station_longitude : the longitude the bike is hired from
* end_station_id : the station id the bike returned to
* end_station_name : the station name the bike to returned to
* end_station_latitude : the station latitude the bike is returned to
* end_station_longitude : the station longitude the bike is returned to
* bike_id : the bike id
* user_type : the user type (casual or member)
* member_birth_year : birth year of the user
* member_gender  : gender of the user
* bike_share_for_all_trip : if bike share was used for all trip u

````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
df1 = pd.read_csv('worked_on')
````

````python
df1.head()
````

## (Visualization 1) Gender Domination

> The Go Fordbike sharing has more males than females, this implies that for consistence revenue generation, efforts should be placed on males and females can be worked on as potentials for revenue growth

````python
# exploring most common gender
color = sb.color_palette()[0];
sb.countplot(data=df1, x='member_gender', color=color)
plt.title('gender domination');
````

# (Visualization 2) Membership type by gender
 Market St at 10th St is the most visited station for bike hiring which indicates the busiest, effort should be put to ensure seemless hiring of bikes 

````python
# slicing or selected the first 10 most visited station 
counts = df1['start_station_name'].value_counts()[0:10]
counts
````

````python
# plotting horinzontal bar chart for the 10 most visited stations
plt.figure(figsize=(10,5));
plt.subplot(1,2,1);
plt.barh(data=counts, y=counts.index, width=counts.values)
plt.xlabel('count')
plt.title('most visited or busiest station');
#plt.x(rotation=90);
````

## (Visualization 3) membership type by gender

> It is interesting to see patronage are more committed and willing to have long term relationship than being just casual users. The plot below shows there are more male subscribers than other genders for both customer membership and subcriber membership. This once again suggest effort should be put on the males for consistency in revenue generation and other gender can be harnessed for revenue growth

````python
sb.countplot(data=df1, x='user_type', hue='member_gender')
plt.xlabel('membership type')
plt.title('membership type by gender');
````

## (Visualization 4) Membership distance travelled

> The plot below shows customer membership(not committed users) are doing longer distance than subscriber membership(committed users). Effort should be made to make non-committed users to committed users to increase profit

````python
plt.figure(figsize=(10,10))
sb.pointplot(data=df1, x='age', y='distance', hue='user_type')
plt.title('membership distance travelled');
````
